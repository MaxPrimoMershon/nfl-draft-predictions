# Will they get drafted?

Goal: Predict whether a wide receiver will get drafted into the NFL.

Process:
1. Data (https://www.reddit.com/r/NFL_Draft/comments/8cxcxj/2018_nfl_draft_data/)
2. Model
3. Evaluate

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import graphviz, pickle
from sklearn import linear_model, metrics, preprocessing, tree
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
pd.set_option('display.max_columns', 500)
```

## 1. Data

```python
data = pd.read_excel(r"C:\Users\Owner\Documents\PERSONAL FILES\Portfolio Website\NFL Draft Predictions\NFL-Draft-2018_Max.xlsx")
print(data.shape)
data.head()
```

```python
data_wr = data[['pick', 'pos', 'year',
       'height_inches', 'forty', 
       'receiving.rec.td', 'receiving.rec.yards',
       'receiving.receptions', 'receiving.seasons', 'Conference']]

data_wr = data_wr.loc[data_wr["pos"] == "WR"]
data_wr = data_wr.loc[data_wr["year"] != 2018]
data_wr = data_wr.loc[data_wr["receiving.seasons"] != 0]

data_wr["receiving.rec.td"] = data_wr["receiving.rec.td"]/data_wr["receiving.seasons"]
data_wr["receiving.rec.yards"] = data_wr["receiving.rec.yards"]/data_wr["receiving.seasons"]
data_wr["receiving.receptions"] = data_wr["receiving.receptions"]/data_wr["receiving.seasons"]

data_wr["Drafted"] = 1
data_wr.loc[data_wr["pick"] == 300, "Drafted"] = 0
data_wr = data_wr.loc[data_wr["forty"].notnull() == True]

data_wr.drop(["pos","year","pick"],inplace=True, axis=1)

print(data_wr.shape)
data_wr.head()
```

```python
data_wr.isnull().sum()
```

```python
data_wr["Drafted"].value_counts()
```

```python
data_wr["Conference"].value_counts()
```

```python
data_wr = pd.get_dummies(data_wr, columns=["Conference"])
data_wr.drop(["Conference_Sun Belt", "Conference_Independent","Conference_MAC","Conference_C-USA","Conference_American","Conference_Mountain West","Conference_Pac-12","Conference_Big 12"],axis=1,inplace=True)
data_wr.head()
```

```python
data_wr.corr()
```

```python
train, test = train_test_split(data_wr, test_size=.3, random_state=42)
x_train = train.drop(["Drafted"], axis=1)
x_test = test.drop(["Drafted"], axis=1)
y_train = train["Drafted"]
y_test = test["Drafted"]

print(len(train), "train +", len(test), "test")
print("X Train: " + str(x_train.shape))
print("Y Train: " + str(y_train.shape))
print("X Test: " + str(x_test.shape))
print("Y Test: " + str(y_test.shape))
```

## 2. Modeling

```python
# Functions to evaluate model
def modelresults(model):
    model.fit(x_train, y_train)
    expected = y_test
    predicted = model.predict(x_test)
    print("Accuracy: "+ str(accuracy_score(expected, predicted)))
    print(confusion_matrix(expected, predicted))  
    print(classification_report(expected, predicted))

def modelaccuracy(model):
    model.fit(x_train, y_train)
    expected = y_test
    predicted = model.predict(x_test)
    accuracy = accuracy_score(expected, predicted)
    return accuracy
```

### Decision Tree

```python
# Compare accuracy of decision trees with different number of depths
accuracy = []
for i in range(1, 11):
    dtree = tree.DecisionTreeClassifier(max_depth=i, random_state=42)
    modelscore = modelaccuracy(dtree)
    accuracy.append(modelscore)
    
plt.figure(figsize=(6, 4))  
plt.plot(range(1, 11), accuracy)
plt.title('Accuracy vs. Tree Depth') 
plt.xlabel('Tree Depth')  
plt.ylabel('Accuracy') 
```

```python
# Accuracy of best decision tree
dtree = tree.DecisionTreeClassifier(max_depth=3, random_state=42)
modelresults(dtree)
```

### Logistic Regression

```python
# Accuracy of logistic regression
logreg = LogisticRegression()
modelresults(logreg)
```

### Neural Network

```python
# Scale data to minimize net's sensitivity to feature scaling
scaler = StandardScaler()
scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# Function to compare mean absolute error of different sized nets
def neuronsize(size):
    mae = []
    layers = [size]

    for i in range(1, 6):   
        mlp = MLPClassifier(random_state=0, hidden_layer_sizes=layers, max_iter=1000000)
        mlp.fit(x_train_scaled, y_train)
        expected = y_test
        predicted = mlp.predict(x_test_scaled)
        modelscore = modelaccuracy(mlp)
        mae.append(modelscore)
        layers.append(size)

    plt.figure(figsize=(6, 3))  
    plt.plot(range(1, 6), mae)
    plt.title('Mean Absolute Error vs. Net Layers (w/ ' + str(size) + " neurons)") 
    plt.xlabel('Net Layers')  
    plt.ylabel('Mean Absolute Error') 
```

```python
# Compare mean absolute error of different # of layers and neurons in each layer
neuronsize(9)
neuronsize(18)
```

```python
mlp = MLPClassifier(random_state=0, hidden_layer_sizes=(18,18), max_iter=1000000)
modelresults(mlp)
```

## 3. Evaluate

```python
# Comparing results of each model
model_scores = pd.DataFrame(index = ["Decision Tree", "Logistic Regression", "Neural Network"], columns=["Accuracy"])

model_scores.at["Decision Tree","Accuracy"] = modelaccuracy(dtree)
model_scores.at["Logistic Regression","Accuracy"] = modelaccuracy(logreg)
model_scores.at["Neural Network","Accuracy"] = modelaccuracy(mlp)

model_scores
```

```python
# Plotting model results
counts = pd.DataFrame(data_wr["Drafted"].value_counts(1))
baseline = counts.at[1, "Drafted"]

model_scores.sort_values("Accuracy", ascending=True).plot.barh(legend=None)
plt.xlabel("Accuracy")
plt.title("Model Accuracy")
plt.axvline(baseline, color='red', linestyle='--')
plt.show()
```
